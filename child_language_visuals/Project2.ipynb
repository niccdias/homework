```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
words_df = pd.read_csv('~/Desktop/Project2/word_data.csv')
```

```python
words_df['growth'] = words_df['30'] - words_df['16']
```

```python
words_filtered = words_df[words_df['growth'] > 0]
```

```python
words_desc = words_filtered.sort(['growth', '30'], ascending=[0, 0])
```

```python
words_desc = words_desc.reset_index()
```

```python
words_desc['category'].unique()
```

```python
words_desc[words_desc['category'] == 'animals'].tail(10)
```

```python
avg_growth_by_category = []

for category in words_desc['category'].unique():
    cat_dict = {}
    df = words_desc[words_desc['category'] == category]
    avg_growth = np.mean(df['growth'])
    cat_dict['category'] = category
    cat_dict['avg_growth'] = avg_growth
    avg_growth_by_category.append(cat_dict)
    
avg_growth_df = pd.DataFrame(avg_growth_by_category)
```

```python
avg_growth_clean = avg_growth_df.sort(['avg_growth'], ascending=[0]).reset_index()
del avg_growth_clean['index']
```

```python
avg_growth_clean.head(10)
```

```python
categories = list(avg_growth_clean['category'][0:10]) + list(avg_growth_clean['category'][-10:])
categories = list(reversed(categories))
y_pos = np.arange(len(categories))
growths = list(avg_growth_clean['avg_growth'][0:10]) + list(avg_growth_clean['avg_growth'][-10:])
growths = list(reversed(growths))

plt.clf()
plt.barh(y_pos, growths)
plt.yticks(y_pos, categories)
plt.ylabel('Type of word')
plt.xlabel('Average growth')
plt.title('Some words are more useful than others')

plt.savefig("growth_by_type.pdf", transparent=True, bbox_inches='tight')
```

```python
child_df = pd.read_csv('~/Desktop/Project2/child_data.csv')
```

```python
child_df.head()
```

```python
child_df['mom_ed'].unique()
```

```python
avg_comp_by_mom_ed = []

for ed in child_df['mom_ed'].unique():
    ed_dict = {}
    df = child_df[child_df['mom_ed'] == ed]
    avg_comp = np.mean(df['comprehension'])
    ed_dict['ed_level'] = ed
    ed_dict['avg_comp'] = avg_comp
    avg_comp_by_mom_ed.append(ed_dict)
    
mom_ed_comp_df = pd.DataFrame(avg_comp_by_mom_ed)
```

```python
mom_comp_clean = mom_ed_comp_df.sort(['avg_comp'], ascending=0).reset_index()
del mom_comp_clean['index']
mom_comp_clean = mom_comp_clean.drop([7])
```

```python
mom_comp_clean
```

```python
y_pos = np.arange(len(mom_comp_clean['ed_level']))
 
plt.clf()    
plt.barh(y_pos, list(reversed(mom_comp_clean['avg_comp'])))
plt.yticks(y_pos, list(reversed(mom_comp_clean['ed_level'])))
plt.ylabel("Mother's level of education")
plt.xlabel("Average number of words understood")
plt.title("Kids with educated moms understand more words")

plt.savefig("mom_comp.pdf", transparent=True, bbox_inches='tight')
```

```python
avg_pro_by_mom_ed = []

for ed in child_df['mom_ed'].unique():
    ed_dict = {}
    df = child_df[child_df['mom_ed'] == ed]
    avg_pro = np.mean(df['production'])
    ed_dict['ed_level'] = ed
    ed_dict['avg_pro'] = avg_pro
    avg_pro_by_mom_ed.append(ed_dict)
    
mom_ed_pro_df = pd.DataFrame(avg_pro_by_mom_ed)
```

```python
mom_pro_clean = mom_ed_pro_df.sort(['avg_pro'], ascending=0).reset_index()
del mom_pro_clean['index']
mom_pro_clean = mom_pro_clean.drop([7])
```

```python
mom_pro_clean
```

```python
y_pos = np.arange(len(mom_pro_clean['ed_level']))
 
plt.clf()    
plt.barh(y_pos, list(reversed(mom_pro_clean['avg_pro'])))
plt.yticks(y_pos, list(reversed(mom_pro_clean['ed_level'])))
plt.ylabel("Mother's level of education")
plt.xlabel("Average number of words used")
plt.title("Kids with educated moms use more words")

plt.savefig("mom_pro.pdf", transparent=True, bbox_inches='tight')
```
